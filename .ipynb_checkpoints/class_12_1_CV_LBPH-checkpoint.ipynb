## ▣ Bottom line  

**Face detection** is the first step for further face analysis, including **recognition, emotion detection, or face generation.**   
- However, it is crucial to collect all the necessary data for further processing.   
- Robust face detection is a prerequisite for sophisticated recognition, tracking, and analytics tools and the cornerstone of computer vision.
    
## ▶ Face detection methods  

These methods are divided into four categories,

<img src='./images/face_det1.png'  width=600 height=400>

**1. Knowledge-based face detection**  
- This method relies on the set of rules developed by humans according to our knowledge.   
- We know that a face must have a nose, eyes, and mouth within certain distances and positions with each other.  
<img src='./images/face_det2.png'  width=600 height=400>
- The problem with this method is to build an appropriate set of rules. 
- If the rules are too general or too detailed, the system ends up with many false positives.   
- And it does not work for all skin colors and depends on lighting conditions that can change the exact hue of a person’s skin in the picture.  

**2. Template matching**  
- The template matching method uses predefined or parameterized face templates to locate or detect the faces by the correlation between the predefined or deformable templates and input images.   
- The face model can be constructed by edges using the edge detection method.
<img src='./images/face_det3.png'  width=600 height=400>
- For this approach, the software has several classifiers for detecting various types of front-on faces and some for profile faces, such as detectors of eyes, a nose, a mouth, and in some cases, even a whole body.   
- While the approach is easy to implement, it is usually inadequate for face detection.  

**3. Feature-based face detection**  
- The feature-based method extracts structural features of the face.   
- It is trained as a classifier and then used to differentiate facial and non-facial regions.   
- One example of this method is **Haar Feature Selection** : 
>- color-based face detection that scans colored images or videos for areas with typical skin color and then looks for face segments.  
>- There are 38 layers of cascaded classifiers to obtain the total number of 6061 features from each frontal face.  
- An other Feature-based face detection example is **HOG(Histogram of Oriented Gradients)**  
>- HOG is a feature extractor for object detection.   
>- The features extracted are the distribution (histograms) of directions of gradients (oriented gradients) of the image.  

<img src='./images/face_det4.png'  width=600 height=400>
- Gradients are typically large round edges and corners and allow us to detect those regions.   
- Instead of considering the pixel intensities, they count the occurrences of gradient vectors to represent the light direction to localize image segments.   
- The method uses overlapping local contrast normalization to improve accuracy.  

**4. Appearance-based face detection**  
- The more advanced appearance-based method depends on a set of delegate training face images to find out face models.   
- It relies on machine learning and statistical analysis to find the relevant characteristics of face images and extract features from them.   
- This method unites several algorithms:  
>- Eigenface-based algorithm  
>- Distribution-based algorithms  
>- Hidden Markov Model  
>- Sparse Network of Winnows 
>- Naive Bayes Classifiers  
>- Inductive learning  
>- Neural networks, such as GANs
>>-  are among the most recent and most powerful methods for detection problems, including face detection, emotion detection, and face recognition.

## ▣ LBPH(Local Binary Patterns Histograms) Algorithm  
- http://www.scholarpedia.org/article/Local_Binary_Patterns  
- https://towardsdatascience.com/face-recognition-how-lbph-works-90ec258c3d6b

__LBPH is one of the easiest face recognition algorithms.__  
- __It can represent local features in the images.__  
- __It is possible to get great results (mainly in a controlled environment).__  
- __It is robust against monotonic gray scale transformations.__  
- __It is provided by the OpenCV library (Open Source Computer Vision Library).__  
- __LBPH is a widely used technique because of its fast computational speed and simple algorithm.__    
- __Utilized in a wide range of applications: facial recognition, license plate recognition, patient diagnostics and biometric authentication, etc.__   
    
> LBPH는 가장 쉬운 얼굴 인식 알고리즘 중 하나입니다.
>- 이미지의 로컬 기능을 나타낼 수 있습니다.
>- (주로 통제된 환경에서) 훌륭한 결과를 얻을 수 있습니다.
>- 단조로운 그레이 스케일 변환에 강합니다.
>- OpenCV 라이브러리(오픈 소스 컴퓨터 비전 라이브러리)에서 제공합니다.
>- LBPH는 빠른 계산 속도와 단순한 알고리즘 때문에 널리 사용되는 기법입니다.  
>- 다양한 응용 분야에서 활용됨: 얼굴 인식, 자동차 번호판 인식, 환자 진단 및 생체 인증 등  
    
Human beings perform face recognition automatically every day and practically with no effort.  
Although it sounds like a very simple task for us, it has proven to be a complex task for a computer, as it has many variables that can impair the accuracy of the methods, for example: illumination variation, low resolution, occlusion, amongst other.  
In computer science, face recognition is basically the task of recognizing a person based on its facial image.   
It has become very popular in the last two decades, mainly because of the new methods developed and the high quality of the current videos/cameras.  

Note that face recognition is different of face detection:

- Face Detection: it has the objective of finding the faces (location and size) in an image and probably extract them to be used by the face recognition algorithm.
- Face Recognition: with the facial images already extracted, cropped, resized and usually converted to grayscale, the face recognition algorithm is responsible for finding characteristics which best describe the image.
<img src='./images/CV/face_recog.png'  width=800 height=400>

The face recognition systems can operate basically in two modes:

- Verification or authentication of a facial image: it basically compares the input facial image with the facial image related to the user which is requiring the authentication. It is basically a 1x1 comparison.
- Identification or facial recognition: it basically compares the input facial image with all facial images from a dataset with the aim to find the user that matches that face. It is basically a 1xN comparison.

Local Binary Pattern (LBP) is a simple yet very efficient texture operator which labels the pixels of an image by thresholding the neighborhood of each pixel and considers the result as a binary number.  
- It was first described in 1994 (LBP) and has since been found to be a powerful feature for texture classification.     
- It has further been determined that when LBP is combined with histograms of oriented gradients (HOG) descriptor, it improves the detection performance considerably on some datasets.  
- Using the LBP combined with histograms we can represent the face images with a simple data vector.  
- As LBP is a visual descriptor it can also be used for face recognition tasks, as can be seen in the following step-by-step explanation.  
  
>인간은 매일 그리고 실질적으로 아무런 노력 없이 자동으로 얼굴 인식을 수행합니다.
>우리에게는 매우 간단한 작업처럼 들리지만 방법의 정확도를 손상시킬 수 있는 많은 변수(예: 조명 변화, 저해상도, 폐색 등)가 있기 때문에 컴퓨터에서는 복잡한 작업임이 입증되었습니다.
>컴퓨터 과학에서 얼굴 인식은 기본적으로 얼굴 이미지를 기반으로 사람을 인식하는 작업입니다. 지난 20년 동안 개발된 새로운 방법과 현재 비디오/카메라의 고품질 때문에 매우 인기를 얻었습니다.
>
>- Face Detection: 이미지에서 얼굴(위치 및 크기)을 찾고 얼굴 인식 알고리즘에서 사용할 수 있도록 추출하는 것을 목표로 합니다.
>- Face Recognition: 얼굴 이미지가 이미 추출, 자르기, 크기 조정되고 일반적으로 회색조로 변환된 상태에서 얼굴 인식 알고리즘은 이미지를 가장 잘 설명하는 특성을 찾는 역할을 합니다.
>
>얼굴 인식 시스템은 기본적으로 두 가지 모드로 작동할 수 있습니다.
>
>- 얼굴 이미지 확인 또는 인증: 기본적으로 입력된 얼굴 이미지를 인증이 필요한 사용자와 관련된 얼굴 이미지와 비교합니다. 기본적으로 1x1 비교입니다.
>- 식별 또는 얼굴 인식: 기본적으로 입력된 얼굴 이미지를 데이터 세트의 모든 얼굴 이미지와 비교하여 해당 얼굴과 일치하는 사용자를 찾습니다. 기본적으로 1xN 비교입니다.

LBPH는 LBP와 히스토그램을 결합한 기법임  
- LBP(Local Binary Patterns)는 영상처리 분야에서 텍스처를 인식하거나 분석하는 데 사용하는 특징 추출 기술입니다.   
- LBP는 이미지의 각 픽셀 주변 픽셀 값을 이진수로 변환해 이미지의 특징을 추출합니다.   
- 이진수는 중심 픽셀과 이웃 픽셀의 상대적인 밝기 차이에 따라 생성됩니다.   
- 이웃 픽셀이 중심 픽셀보다 크면 1, 작으면 0으로 표현하죠. 
- 이렇게 만든 이진수를 활용해 각 픽셀의 텍스처 특징을 추출할 수 있습니다.   
- 국지 이진  패턴(Local Binary Patterns, LBP)을 추출하면 이미지의 특정 부분에서 특정한 패턴이 나타날 수 있습니다.   
- 이러한 패턴을 활용해 이미지에서 물체나 얼굴을 인식하는 겁니다.  

>LBP(Local Binary Pattern)는 각 픽셀의 이웃을 임계값으로 설정하여 이미지의 픽셀에 레이블을 지정하고 결과를 이진수로 간주하는 간단하지만 매우 효율적인 텍스처 연산자입니다.  
>1994년(LBP)에 처음 기술되었으며 이후 텍스처 분류를 위한 강력한 기능으로 밝혀졌습니다.    
>- LBP가 HOG(Histograms of Oriented Gradients) 디스크립터와 결합되면 일부 데이터 세트에서 검출 성능이 상당히 향상되는 것으로 확인되었습니다.  
>- 히스토그램과 결합된 LBP를 사용하여 간단한 데이터 벡터로 얼굴 이미지를 표현할 수 있습니다.  
>- LBP는 시각적 설명자이므로 다음 단계별 설명에서 볼 수 있듯이 얼굴 인식 작업에도 사용할 수 있습니다.  

## ▶ Step-by-Step

**1. LBPH uses 4 parameters:**  
      
- Radius: the radius is used to build the circular local binary pattern and represents the radius around the central pixel. It is usually set to 1.  
- Neighbors: the number of sample points to build the circular local binary pattern. Keep in mind: the more sample points you include, the higher the computational cost. It is usually set to 8.  
- Grid X: the number of cells in the horizontal direction. The more cells, the finer the grid, the higher the dimensionality of the resulting feature vector. It is usually set to 8.  
- Grid Y: the number of cells in the vertical direction. The more cells, the finer the grid, the higher the dimensionality of the resulting feature vector. It is usually set to 8.  

>- 매개변수: LBPH는 4개의 매개변수를 사용합니다.
>- 반지름: 반지름은 원형 로컬 이진 패턴을 만드는 데 사용되며 중앙 픽셀 주위의 반지름을 나타냅니다. 일반적으로 1로 설정됩니다.
>- Neighbors: 순환 로컬 이진 패턴을 구축하기 위한 샘플 점의 수입니다. 포함되는 샘플 점이 많을수록 계산 비용이 높아집니다. 일반적으로 8로 설정됩니다.
>- 그리드 X: 가로 방향의 셀 수입니다. 셀이 많을수록 그리드가 더 미세하고 결과 기능 벡터의 차원이 높아집니다. 일반적으로 8로 설정됩니다.
>- 그리드 Y: 세로 방향의 셀 수입니다. 셀이 많을수록 그리드가 더 미세하고 결과 기능 벡터의 차원이 높아집니다. 일반적으로 8로 설정됩니다.    

>- Neighbors(이웃 픽셀 수) : LBP를 만들 때 사용할 이웃 픽셀 수를 뜻합니다. 이웃 픽셀 수가 많을수록 계산 비용이 높아집니다. 보통 이 값은 8로 설정합니다.
>- Grid X(수평 방향 분할 수) : 수평 방향으로 셀을 분할할 개수를 말합니다. 보통 8로 설정합니다.
>- Grid Y(수직 방향 분할 수) : 수직 방향으로 셀을 분할할 개수를 말합니다. 보통 8로 설정합니다.  
 
**2. 2. 데이터 준비  Training the Algorithm:**   
    
- First, we need to train the algorithm.   
- To do so, we need to use a dataset with the facial images of the people we want to recognize.   
- We need to also set an ID (it may be a number or the name of the person) for each image, so the algorithm will use this information to recognize an input image and give you an output. Images of the same person must have the same ID.   

>- 이렇게 하려면 인식하려는 사람들의 얼굴 이미지가 포함된 데이터 세트를 사용해야 합니다. 
>- 또한 각 이미지에 대해 ID(숫자 또는 사람 이름일 수 있음)를 설정해야 알고리즘이 이 정보를 사용하여 입력 이미지를 인식하고 출력을 제공합니다. 동일 인물의 이미지는 동일한 ID를 가져야 합니다. 

>- 얼굴 인식에 LBPH를 활용하려면 먼저 LBPH 알고리즘을 데이터로 훈련해야겠죠? 
>- 인식하려는 사람들로 이루어진 데이터를 활용해서요.   
>- 각 얼굴마다 고유한 ID도 있어야 합니다. 그 얼굴이 누구를 나타내는지도 중요하기 때문이죠.   
>- 그래야 개별 얼굴이 누구를 나타내는지 분간해서 훈련을 할 수 있습니다.   
>- 서로 다른 이미지에 같은 사람이 있다면 동일한 ID를 가져야 합니다.  

**3. Applying the LBP operation:**   
    
- The first computational step of the LBPH is to create an intermediate image that describes the original image in a better way, by highlighting the facial characteristics.   
- To do so, the algorithm uses a concept of a sliding window, based on the parameters radius and neighbors.
>- LBPH의 첫 번째 계산 단계는 얼굴 특성을 강조하여 원본 이미지를 더 나은 방식으로 설명하는 중간 이미지를 만드는 것입니다.   
>- 이를 위해 알고리즘은 매개 변수 radius 및 neighbors를 기반으로 하는 슬라이딩 윈도우 개념을 사용합니다.  
- It's time to work on extracting the LBP to save the characteristic areas from the original image.   
- This means that you need to replace the original image with the LBP extracted image.   
- Creating an LBP image allows you to highlight the features on your face that you should pay attention to. 
- Use the Neighbors parameter here.
>- 원본 이미지에서 특징적인 영역을 구하기 위해 LBP를 추출하는 작업을 할 차례입니다.   
>- 원본 이미지를 LBP가 추출된 이미지로 바꿔야 한다는 말입니다.   
>- LBP 이미지를 만들면 얼굴에서 주목해야 할 특징을 강조할 수 있습니다. 
>- 이때 Neighbors 파라미터를 사용합니다. 

The image below shows this procedure:

<img src='./images/CV/lbph2.png'  width=800 height=400>

Based on the image above, let’s break it into several small steps so we can understand it easily:

- Suppose we have a facial image in grayscale.  
- We can get part of this image as a window of 3x3 pixels.  
- It can also be represented as a 3x3 matrix containing the intensity of each pixel (0~255).  
- Then, we need to take the central value of the matrix to be used as the threshold.  
- This value will be used to define the new values from the 8 neighbors.  
- For each neighbor of the central value (threshold), we set a new binary value. We set 1 for values equal or higher than the threshold and 0 for values lower than the threshold.  
- Now, the matrix will contain only binary values (ignoring the central value). 
>- We need to concatenate each binary value from each position from the matrix line by line into a new binary value (e.g. 10001101).   
>- Note: some authors use other approaches to concatenate the binary values (e.g. clockwise direction), but the final result will be the same.  
<img src='./images/CV/lbph3_1.png'  width=400 height=200>

- Then, we convert this binary value to a decimal value and set it to the central value of the matrix, which is actually a pixel from the original image.  
- At the end of this procedure (LBP procedure), we have a new image which represents better the characteristics of the original image.  
<img src='./images/CV/lbph3.png'  width=300 height=300>

- Note: The LBP procedure was expanded to use a different number of radius and neighbors, it is called **Circular LBP.**  

<img src='./images/CV/lbph6.png'  width=600 height=400>
<center>Circular LBP</center>

>- It can be done by using bilinear interpolation.   
>- If some data point is between the pixels, it uses the values from the 4 nearest pixels (2x2) to estimate the value of the new data point.

>- 회색조의 얼굴 이미지가 있다고 가정합니다.
>- 이 이미지의 일부를 3x3 픽셀의 창으로 가져올 수 있습니다.
>- 각 픽셀의 강도(3~3)를 포함하는 0x255 행렬로 나타낼 수도 있습니다.
>- 그런 다음 임계값으로 사용할 행렬의 중심 값을 가져와야 합니다.
>- 이 값은 8개의 인접 항목에서 새 값을 정의하는 데 사용됩니다.
>- 중심 값(임계값)의 각 이웃에 대해 새 이진 값을 설정합니다. 
>- 임계값보다 크거나 같은 값에 대해 1을 설정하고 임계값보다 낮은 값에 대해 0을 설정합니다.
>- 이제 행렬에는 이진 값만 포함됩니다(중심 값은 무시).   
>- 행렬의 각 위치에서 각 이진 값을 한 줄씩 새 이진 값(예: 10001101)으로 연결해야 합니다.   
>- 참고: 일부 작성자는 다른 접근 방식을 사용하여 이진 값(예: 시계 방향)을 연결하지만 최종 결과는 동일합니다.
>- 그런 다음 이 이진 값을 10진수 값으로 변환하고 실제로 원본 이미지의 픽셀인 행렬의 중심 값으로 설정합니다.
>- 이 절차(LBP 절차)가 끝나면 원본 이미지의 특성을 더 잘 나타내는 새 이미지가 있습니다.
>- 참고: LBP 절차는 다른 수의 반지름 및 인접 항목을 사용하도록 확장되었으며, 이를 원형 LBP라고 합니다.
>>- 쌍선형 보간을 사용하여 수행할 수 있습니다. 
>>- 일부 데이터 포인트가 픽셀 사이에 있는 경우 가장 가까운 픽셀 4개(2x2)의 값을 사용하여 새 데이터 포인트의 값을 추정합니다.

맨 왼쪽이 원본 이미지(흑백 이미지라고 가정)입니다.   
먼저, 원본 이미지에서 3x3 픽셀로 이루어진 윈도(window)를 뽑습니다.   
흑백 이미지라서 각 픽셀은 0~255 사이의 값 하나로 이루어졌겠죠. 
픽셀 값이 밝기를 나타낸다고 봐도 됩니다.   
- 0이면 완전 검은색, 255면 완전 흰색을 뜻하니까요.   
- 3x3 픽셀로 이루어진 윈도의 가운데값을 임계값(threshold)으로 정합니다. 여기서는 임계값이 90이 됩니다.   
- 다음으로 이웃 픽셀값을 순회하면서 임계값보다 작으면 0, 크면 1로 값을 매깁니다.   
- 그러면 이웃 픽셀은 모두 0 또는 1로 설정되겠죠.   
- 임계값인 가운데 값은 빼고 나머지 이웃 셀의 이진 값을 활용해 이진수를 만듭니다. 여기서는 10001101입니다. 왼쪽에서 오른쪽, 위에서 아래로 훑으며 이진 값을 뽑아서 10001101이 된 겁니다. 그런데 시계방향으로 훑어도 됩니다. 가령, 10011010이 되어도 됩니다. 최종 결과에는 변함이 없어서 어떤 방식으로 이진 값 순서를 매겨도 무방합니다. 그다음은 10001101을 십진수로 바꿉니다. 141이 되겠죠. 마지막으로, 가운데 픽셀 값을 141로 바꿉니다. 

이런 방식으로 LBP 이미지를 만들면 아래와 같은 결과를 얻게 됩니다.

**4. Extracting the Histograms:**   
    
- Now, using the image generated in the last step, we can use the Grid X and Grid Y parameters to divide the image into multiple grids, as can be seen in the following image:

<img src='./images/CV/lbph4.png'  width=800 height=400>

Based on the image above, we can extract the histogram of each region as follows:

- As we have an image in grayscale, each histogram (from each grid) will contain only 256 positions (0~255) representing the occurrences of each pixel intensity.
- Then, we need to concatenate each histogram to create a new and bigger histogram.   
- Supposing we have 8x8 grids, we will have 8x8x256=16,384 positions in the final histogram.   
- The final histogram represents the characteristics of the image original image.
>- 이제 마지막 단계에서 생성된 이미지를 사용하여 다음 이미지에서 볼 수 있듯이 앞서 만든 LBP 이미지를 (Grid X, Grid Y)로 분할합니다. 
>- 그리드 X 및 그리드 Y 매개변수를 사용하여 이미지를 여러 그리드로 나눌 수 있습니다.
>- 일반적으로 두 값 모두 8로 잡는다고 했습니다. 여기서도 LBP 이미지를 8x8 그리드로 나눴습니다.
>- 위의 이미지를 기반으로 다음과 같이 각 영역의 각 그리드별로 픽셀 값의 크기에 따라 히스토그램을 추출할 수 있습니다.
>- 회색조의 이미지가 있으므로 각 그리드의 각 히스토그램에는 각 픽셀 강도의 발생을 나타내는 256개의 위치(0~255)만 포함됩니다.
>- 그런 다음 각 히스토그램을 연결하여 새롭고 더 큰 히스토그램을 만들어야 합니다.   
>- 8x8 그리드가 있다고 가정하면 최종 히스토그램에 8x8x256=16,384 위치가 있습니다.   
>- 최종 히스토그램은 이미지 원본 이미지의 특성을 나타냅니다.   

So you will get a total of 64 histograms (=8x8), right?   
Combine 64 histograms to create a final histogram.   
This histogram is unique to that face.   
If it's the same person, the histogram pattern will be similar. If they are different people, the histogram will be different.   
The histogram pattern allows you to tell if it is the same person or someone else.   
The larger the number of Gird X and Grid Y, the higher the number of grids.   
This makes the cells smaller, so the total number of histograms increases.   
More histograms means more patterns to be extracted.

So how do you determine if it's a similar histogram?   
There are several ways to do this.   
Examples include Euclidean distances and absolute distance calculations.   
If this value is less than a certain threshold, it is considered a histogram from the same image.   
Thus, they perceive them as the same person. Depending on the threshold, the quality of the result will vary.   

그럼 히스토그램이 총 64개(=8x8) 나오겠죠?   
64개 히스토그램을 결합해 최종 히스토그램을 만듭니다.   
이 히스토그램이 해당 얼굴에 대한 고유한 히스토그램입니다.   
같은 사람이라면 히스토그램 패턴이 비슷할 겁니다. 서로 다른 사람이라면 히스토그램도 달라지겠죠.   
히스토그램 패턴을 통해 동일한 사람인지 다른 사람인지 분간을 할 수 있습니다.   
Gird X와 Grid Y 개수가 크면 그만큼 그리드 개수도 많아집니다.   
그러면 셀이 작아지므로 전체 히스토그램 개수도 많아지죠.   
히스토그램이 많다는 말은 더 많은 패턴을 추출한다는 뜻입니다.

그럼 비슷한 히스토그램인지는 어떻게 판단할까요?   
여러 방법이 있습니다.   
대표적으로 유클리디안 거리나 절댓값 거리 계산법이 있습니다.   
이 값이 특정한 임계값보다 작다면 동일한 이미지에서 나온 히스토그램으로 간주하는 거죠.   
그리하여 같은 사람으로 인식합니다. 임계값에 따라 결과 품질이 달라집니다.  

**5. Performing the face recognition:**   
    
- In this step, the algorithm is already trained.   
- Each histogram created is used to represent each image from the training dataset.   
- So, given an input image, we perform the steps again for this new image and creates a histogram which represents the image.

- So to find the image that matches the input image we just need to compare two histograms and return the image with the closest histogram.  

- We can use various approaches to compare the histograms (calculate the distance between two histograms),   
- for example: euclidean distance, chi-square, absolute value, etc.   
- In this example, we can use the Euclidean distance (which is quite known) based on the following formula:
<center>$ D = \sqrt{\sum_{i=1}^N (hist1_i - hist2_i)^2}$ </center>
- So the algorithm output is the ID from the image with the closest histogram.   
- The algorithm should also return the calculated distance, which can be used as a‘confidence’measurement.   
- Note: the confidence, as lower confidences are better because it means the distance between the two histograms is closer.
- We can then use a threshold and the confidence to automatically estimate if the algorithm has correctly recognized the image.   
- We can assume that the algorithm has successfully recognized if the confidence is lower than the threshold defined.
>- 이 단계에서는 알고리즘이 이미 학습되어 있습니다. 생성된 각 히스토그램은 학습 데이터 세트의 각 이미지를 나타내는 데 사용됩니다. 따라서 입력 이미지가 주어지면 이 새 이미지에 대해 단계를 다시 수행하고 이미지를 나타내는 히스토그램을 만듭니다.
>- 따라서 입력 이미지와 일치하는 이미지를 찾으려면 두 히스토그램을 비교하고 가장 가까운 히스토그램이 있는 이미지를 반환하기만 하면 됩니다.
>- 히스토그램을 비교하기 위해 다양한 접근 방식을 사용할 수 있습니다(두 히스토그램 사이의 거리 계산),   
>- 예를 들어 유클리드 거리, 카이제곱, 절대값 등. 이 예에서는 다음 공식을 기반으로 유클리드 거리(잘 알려져 있음)를 사용할 수 있습니다.
>- 따라서 알고리즘 출력은 가장 가까운 히스토그램이 있는 이미지의 ID입니다. 알고리즘은 또한 '신뢰도' 측정으로 사용할 수 있는 계산된 거리를 반환해야 합니다.   
>- 참고: 신뢰도가 낮을수록 두 히스토그램 사이의 거리가 더 가깝다는 것을 의미
>- 그런 다음 임계값과 '신뢰도'를 사용하여 알고리즘이 이미지를 올바르게 인식했는지 자동으로 추정할 수 있습니다. 신뢰도가 정의된 임계값보다 낮은지 알고리즘이 성공적으로 인식했다고 가정할 수 있습니다.

```python
import zipfile

fer_path = './images/CV/Yale_Face.zip'
zip = zipfile.ZipFile(file=fer_path, mode='r')
zip.extractall('./images/CV/Yale_Face/')  # extract all file in current directory
zip.close()
```

```python
import os
yalefaces_path = './images/CV/Yale_Face'
all_image_file_name = os.listdir(yalefaces_path)
print(all_image_file_name)
```

```python
import os
train_image_path = './images/CV/Yale_Face/train'
train_f = os.listdir(train_image_path) # print all train image file name
print(train_f)
```

```python
import numpy as np
from PIL import Image

def fetch_preprocessed_image_data(image_paths):
    faces_of_image = []
    ids_of_image = []

    for image_path in image_paths:
        image = Image.open(image_path).convert('L')  # convert to grayscale
        # convert the image type to a numpy array type (with integer format)
        image_np = np.array(image, 'uint8')  
        image_file_name = os.path.split(image_path)[1]
        id_of_image = int(image_file_name.split('.')[0].replace('subject', ''))
        ids_of_image.append(id_of_image)
        faces_of_image.append(image_np)

    return np.array(ids_of_image), faces_of_image
```

```python
ids_of_train_image, faces_of_train_image = fetch_preprocessed_image_data(train_f)
```

```python
ids_of_train_image
```

```python
faces_of_train_image[0]
```

```python
len(ids_of_train_image), len(faces_of_train_image)
```

```python
# LBP 
from skimage.feature import local_binary_pattern

original_image = faces_of_train_image[5] # Original image as a numpy array
lbp_image = local_binary_pattern(original_image, 8*3, 3) # Compute LBP image with radius=3 and n_points=8*3
lbp_image = ((lbp_image - lbp_image.min()) / (lbp_image.max() - lbp_image.min())) * 255 # Normalize
lbp_image = lbp_image.astype(np.uint8) # Convert the LBP image to integer data type
```

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(10, 5), dpi=100)
plt.subplot(121),plt.imshow(original_image, 'gray'),plt.title('Org_Img'),plt.axis('off')
plt.subplot(122),plt.imshow(lbp_image, 'gray'),plt.title('LBP_Img'),plt.axis('off')
plt.show()
```

```python
# LBPH FaceRecognizer
import cv2

lbph_recognizer = cv2.face.LBPHFaceRecognizer_create(radius=5, neighbors=15, grid_x=10, grid_y=10)
lbph_recognizer.train(faces_of_train_image, ids_of_train_image)  # Train the lbph recognizer  

test_image_path = './images/CV/Yale_Face/test'
all_test_image_file_name = os.listdir(test_image_path)
all_test_image_paths = [os.path.join(test_image_path, f) for f in all_test_image_file_name]
ids_of_test_image, faces_of_test_image = fetch_preprocessed_image_data(all_test_image_paths)

id_pred, confidence_pred = lbph_recognizer.predict(faces_of_test_image[0])
id_pred, confidence_pred
# (11, 88.48699892288407) id_pred, confidence_pred

cv2.putText(faces_of_test_image[0], f'Pred: {str(id_pred)}', (5, 30), cv2.FONT_HERSHEY_PLAIN, 1.1, (0, 255, 0))
cv2.putText(faces_of_test_image[0], f'True: {str(ids_of_test_image[0])}', (5, 60), cv2.FONT_HERSHEY_PLAIN, 1.1, (0,255,0))

plt.imshow(faces_of_test_image[0], 'gray')
plt.show()
```

```python
# Evaluation

ids_pred = []
for face_of_test_image in faces_of_test_image:
    id_pred, _ = lbph_recognizer.predict(face_of_test_image)
    ids_pred.append(id_pred)

from sklearn.metrics import accuracy_score

accuracy_score(ids_of_test_image, ids_pred)
```
